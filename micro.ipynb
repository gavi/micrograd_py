```python
from micrograd import Value
from graphviz import *
```

```python
x  = Value(10,label = 'x')
y = Value(20, label = 'y')

z = 2*x**2 + 3 *y ; z.label ='z'
z.grad = 1.0

def draw_dot(x):
    graph = Digraph("Micrograd")
    def fill(n, graph):
        graph.node(str(id(n)),f'{n.label}|{n.data}|{n.grad}|{n.operator}',shape = 'record')
        for c in n.came_from:
            fill(c, graph)
            graph.edge(str(id(c)),str(id(n)))
    fill(x, graph)
    return graph

draw_dot(z)
```

```python
z.backward()
draw_dot(z)
```
